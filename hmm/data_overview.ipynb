```python
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
import os
import pandas as pd
```

```python
df = pd.read_csv("/zhouyuyang/project/ProteinFeatureEngineering/data/all.csv")
df['number'] = None
df.columns
```

```python
family_name = None
protein_name = None
records = None
cnt = 0
protein_cnt = 0
for _, row in df.iterrows():
    if type(row['FP']) is not str:
        continue
    if type(row['protein name']) is str:
        protein_name = row['protein name']
        protein_cnt = 0
    if type(row['Familiy/genus']) is str and row['Familiy/genus'] != family_name:
        if records is not None:
            SeqIO.write(records, f"/zhouyuyang/project/ProteinFeatureEngineering/data/fp/{family_name}.fasta", "fasta")
            cnt += 1
            print(f"write {family_name}.fasta, cnt = {cnt}, protein_cnt = {len(records)}")
        family_name = row['Familiy/genus'].replace(' ', '-')
        protein_cnt += 1
        records = [SeqRecord(Seq(row['FP']), id=protein_name+'_'+str(protein_cnt), description=row['Familiy/genus'])]
    else:
        protein_cnt += 1
        records.append(SeqRecord(Seq(row['FP']), id=protein_name+'_'+str(protein_cnt), description=family_name))
```

```python
fasta_root = "/zhouyuyang/project/ProteinFeatureEngineering/data/fasta/"

for file in os.listdir(fasta_root):
    path = fasta_root + file
    records = SeqIO.parse(path, "fasta")
    # print(f"Sequence number: {len(list(records))}")
    # print(file)
    
    df.loc[df["file name"]==file, 'number'] = len(list(records))
df.to_csv("/zhouyuyang/project/ProteinFeatureEngineering/data/all.csv", index=False)
```
